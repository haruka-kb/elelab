## Gitとは？
**Git**(ギット)とはプロジェクト、リソースをクラウドとして効率よく管理するためのツールです<br>
本来は**Git Server**を自分で構築してコマンドライン基準で操作を行うものなのですが、それを便利にしてくれるクラウドリソース、**Github**が存在します

実はこの教材もローカルで作成したテキストをGithubを使って閲覧できるように管理しています

## アカウント取得
[**Github**](https://github.com/)にてアカウントを取得してください。アカウント取れない人がいないとエレラボ一同は信じております

## Githubデスクトップ取得
[**Github Desktop**](https://desktop.github.com/)にてGithub Desktopをダウンロード、インストールしましょう<br>
Github Desktopはデスクトップ側で発行するコマンドを全て代わりにやってくれるGUIツールです<br>

## レポジトリ作成
レポジトリとはプロジェクト・リソースの塊であり、管理単位です。まずはこれを作成しない事にはGit活用は始まりません

### Webから作成
まずはGithubのトップページにアクセスし、右上の**＋**から**New Repository**を選択しましょう
<img src="git_01.png" width="200" height="200">

次に作成画面です
<img src="git_02.png" width="480" height="480">

1. **レポジトリ名**
    - プロジェクト名と一致されるのが一般的です
2. **レポジトリ説明**
    - どのようなレポジトリなのかの説明を短くまとめておきましょう
3. **ローカリティ**
    - とても重要なレポジトリのローカリティです
    - 現在は無料ユーザーもプライベートレポジトリを作成できるようになりました
    - **パブリック:** 全ての訪問者が内容を閲覧可能です。コミットを送信できる人は選べますが、誰でもコメントは残せます
    - **プライベート:** 閲覧可能なユーザー、コミットを送信できるユーザーなどそれぞれ対象を選択可能になります
4. **Initialize with README.MD**
    - チェックを入れた場合、マークダウン形式のレドメが作成され、サマリとしてレポジトリのトップぺージで確認できるようになります
5. **gitignore**
    - 初心者にはあまり関係ありません
    - gitignoreではハイライトなどで無視するファイル形式、ディレクトリを指定することができます
6. **license**
    - 初心者にはあまり関係ありません
    - レポジトリが使用しているライセンスを指定することができます
    - レポジトリが発行するライセンスではないので注意しましょう
    
全ての項目は後から編集可能ですが、**ローカリティ**の項目は特に注意して設定しましょう<br>
設定を終えたのちに**Create Repository**をクリックすればレポジトリ製作完了です

### デスクトップアプリケーションから作成
GUIの左上に存在するFileタブからNew Repositoryを選択すると以下のような画面が出てきます
<img src="git_03.png" width="320" height="480">
入力する内容はWebから作成する場合と同じなので省略します

デスクトップから作成した場合は以下の課程、**クローン**を省略することが可能です

## レポジトリのクローン
Web側に存在するレポジトリを個人コンピュータ側で編集するには**クローン**と呼ばれる作業を行う必要があります

クローンとはクラウド上に存在するデータをデスクトップ側へ完全コピーすることです

### クローン方法
まず、Github Desktopの右上の**File**タブから**Clone Repository**を選択します
<img src="git_04.png" width="200" height="200">

次に、以下のような画面が表示されるのでクローンしたいレポジトリを選択して**Clone**を押します
<img src="git_05.png" width="320" height="480">

これでクライアント側にクローンが作成されました！<bR>
ここまできてやっと編集を開始することが可能です

## コミットの作成と送信
レポジトリに適応した変更をGitへ共有するには**コミット**を作成して送信する必要があります<br>
コミットを送ることでデスクトップ側の変更点がクラウドに適応され、アップデートされます

## コラボレーター
1つのレポジトリを複数人で編集するためには、管理者が**コラボレーター**を追加する必要があります

コラボレーター登録されたユーザーが該当レポジトリに対して得る権限は以下の通りです
- 閲覧
- クローン
- コミット
- ブランチ作成
- プルリクエスト
- マージ承認
- イシュー・プロジェクトへの書き込み

## ブランチング
### ブランチとは
ブランチとは特定の作業を行う為に、一度メインから作業空間を分離することです<br>
ブランチは簡易的に分離されたレポジトリと考えることも可能です

アプリケーションを開発する際など、ブランチを分けて作業を行う事は非常に重要になります<br>
